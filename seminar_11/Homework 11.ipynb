## Нахождение экстремумов и интервалов убывания и возрастания

```python
from sympy.abc import x, y
from sympy import *

x1, x2 = -3, 3 # Задаём интервал
xr = [x/100 for x in range(x1*100, x2*100 + 1)] # Здесь также определяем точность по x до сотых
yr = [func.subs(x, value) for value in xs]
points = list(zip(xr, yr))
# points
# points.sort(key = lambda x: (x[1]))
# extrem_min = points[0]
# extrem_max = points[-1]
# print(f'Минимальный экстремум - {extrem_min}. Максимальный экстремум - {extrem_max}')
extrems = []
cur_x = points[0]
decreasing_mark = True
for item in points[1:]:
    if (decreasing_mark and item[1] < cur_x[1]) or (not decreasing_mark and item[1] > cur_x[1]):
        cur_x = item
        continue
    else:
        extrems.append(cur_x)
        cur_x = item
        decreasing_mark = not decreasing_mark
print(f'Точки экстремума - {extrems}')
print(f"В промежутке от -3 до 3 функция убывает на интервале от -3 до {extrems[0][0]} и от {extrems[1][0]} до {extrems[2][0]}. Возрастает на интервале от {extrems[0][0]} до {extrems[1][0]}")
        
```
